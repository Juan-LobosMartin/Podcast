```python
podcast_data = pd.DataFrame(columns=['Program','# progam','day','# week day','Time','# on that hour','mean','std'])
```

```python
podcast_data
```

```python
# Pod_My_Ma[(Pod_My_Ma['Publication_Day']== 'Saturday') & (Pod_My_Ma['Publication_Time']== 'Morning')]['Episode_Length_minutes'].median()
```

```python
n=0
for podcast in Podcast_Test_df['Podcast_Name'].unique():
    podname_df = Podcast_Train_df[Podcast_Train_df['Podcast_Name'] == podcast]
    feature = podcast
    program_num = len(podname_df)
    
    for day in podname_df['Publication_Day'].unique():
        pub_day_df = podname_df[podname_df['Publication_Day'] == day ]
        pod_day = day
        pod_day_num = len(pub_day_df)

        for hour in pub_day_df['Publication_Time'].unique():
            pub_hour_df = pub_day_df[pub_day_df['Publication_Time'] == hour ]
            time = hour
            time_num = len(pub_hour_df)
            mean = pub_hour_df['Episode_Length_minutes'].mean()
            std = pub_hour_df['Episode_Length_minutes'].std()
            podcast_data.loc[n] = [feature, program_num, pod_day, pod_day_num, time, time_num, mean, std]
            n = n +1
        n= n +4
    n=n +7
        
```

```python
podcast_data
```

```python
podcast_data.to_csv('Data/Podcast_data.csv', index=True)
```
